```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime, timedelta
import math
import numpy as np
```

```python
# 1. Cargar el CSV
CALFILE = '/home/jmrobles/Podcasts/Coffee Break/CB - Hoja de ayuda.csv'
df = pd.read_csv(CALFILE, 
                 header=None, 
                 delimiter=';', 
                 names=['fecha', 'ep', 'gen1', 'gen2', 'gen3'])
df['fecha'] = pd.to_datetime(
    df['fecha'],
    format='%d/%m/%y',
)
print(df.dtypes)
```

```python
fechas = (
    df['fecha']
    .drop_duplicates()
    .sort_values()
    .reset_index(drop=True)
)

new_df = pd.DataFrame({
    'fecha': fechas,
    'episode': fechas.index + 1  # índice 0->1, 1->2, ..., n->n+1
})

# Comprobación de tipos y contenidos
print(new_df.dtypes)
print(new_df.head())
```
